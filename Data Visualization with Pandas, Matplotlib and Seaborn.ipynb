### Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 15, linestyle = '--', markeredgecolor = 'blue')

plt.title('Matplotlib Viz', fontdict = {'fontname' : 'Times New Roman', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```

```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 15, linestyle = '--', markeredgecolor = 'blue')

z = [1,5,10]
plt.plot(x,z, label = 'Baths', color = 'red', marker = '.', markersize = 25, linestyle = '--', markeredgecolor = 'green')


plt.title('Matplotlib Viz', fontdict = {'fontname' : 'Times New Roman', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```

```python
x = [1,2,3]
y = [5,10,20]

plt.figure(figsize = (20,5))

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 15, linestyle = '--', markeredgecolor = 'blue')

z = [1,5,10]
plt.plot(x,z, label = 'Baths', color = 'red', marker = '.', markersize = 25, linestyle = '--', markeredgecolor = 'green')


plt.title('Matplotlib Viz', fontdict = {'fontname' : 'Times New Roman', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.savefig(r'/Users/akhilnashaboina/Documents/Pandas for Data Analysis/matplotlib.png')

plt.show()
```

```python
smokers = smoking[smoking['smoke'] == 'Yes']
```

```python
fig, axs = plt.subplots(1,2)

sns.histplot(female_smokers['age'], bins = 25, ax = axs[0])
axs[0].set_title('Age Distribution - Female Smokers')

sns.histplot(male_smokers['age'], bins = 25, ax = axs[1])
axs[1].set_title('Age Distribution - Male Smokers')


plt.show()
```

```python
housing = pd.read_csv('/Users/akhilnashaboina/Documents/Pandas for Data Analysis/real_estate.csv')
```

```python
housing.head()
```

```python
housing_filt = housing.drop(columns=['status', 'city', 'state', 'zip_code', 'prev_sold_date'], axis = 1)
```

```python
correlation = housing_filt.corr()
```

```python
sns.heatmap(correlation, annot = True, cmap = 'coolwarm')
```

```python
pearson_correlation = housing_filt.corr(method = 'pearson')
pearson_correlation
```

```python
spearman_correlation = housing_filt.corr(method = 'spearman')
spearman_correlation
```

```python
kendall_correlation = housing_filt.corr(method = 'kendall')
kendall_correlation
```

```python
sns.heatmap(pearson_correlation, annot = True, cmap = 'coolwarm')
plt.title('Pearson Correlation')
plt.show()
```

```python
sns.heatmap(spearman_correlation, annot = True, cmap = 'coolwarm')
plt.title('Pearson spearman_correlation')
plt.show()
```

```python
sns.heatmap(kendall_correlation, annot = True, cmap = 'coolwarm')
plt.title('kendall_correlation Correlation')
plt.show()
```

```python
fig, axs = plt.subplots(3,1, figsize = (5,10))

sns.heatmap(pearson_correlation, annot = True, cmap = 'coolwarm', ax = axs[0])
axs[0].set_title('Pearson Correlation')

sns.heatmap(spearman_correlation, annot = True, cmap = 'coolwarm', ax = axs[1])
axs[1].set_title('spearman_correlation')

sns.heatmap(kendall_correlation, annot = True, cmap = 'coolwarm', ax = axs[2])
axs[2].set_title('kendall_correlation')

plt.show()
```
